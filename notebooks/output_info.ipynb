```python
import pandas as pd
import matplotlib.pyplot as plt
import lxml.etree as ET
import xml.etree.ElementTree as ET
#import ElementTree as ET
```

### Emission Output Basic Model

```python
df_cols = ["time","id", "eclass", "CO2", "CO", "HC", "PMx", "NOx","PMx", "fuel","electricity", "route" ,"type" ,"noise", "waiting", "lane" , "pos" , "speed", "angle", "x", "route", "y"]
```

```python
def parse_XML(xml_file, df_cols): 
    """Parse the input XML file and store the result in a pandas 
    DataFrame with the given columns. 
    """
    out_df = pd.DataFrame(columns = df_cols)
    xtree = ET.parse(xml_file)
    xroot = xtree.getroot()

    
    for node in xroot: 
        rows = []
        j = 0
        for element in node:
            res = []
            res.append(node.attrib.get(df_cols[0]))
            
            for el in df_cols[1:]:
                if node is not None and element.get(el) is not None:
                    res.append(element.get(el))
                else: 
                    res.append(None)
            rows.append({df_cols[i]: res[i] 
                        for i, _ in enumerate(df_cols)})
    
            out_df.loc[len(out_df)] = rows[j] 
            j += 1
        
        #print(node.attrib.get(df_cols[0]))
        
    return out_df
```

```python
def parse_xml_to_dataframe(xml_file, columns):
    """
    Parses an XML file and returns a pandas DataFrame with the specified columns, including the root element as a time column.
    
    Args:
    - xml_file: str, the path of the XML file to parse.
    - columns: list of str, the names of the columns to include in the DataFrame.
    
    Returns:
    - df: pandas DataFrame, the parsed data in a DataFrame with the specified columns and a time column for the root element.
    """
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()
    
    # Extract data from XML and store it in a list of dictionaries
    data = []
    i = 0
    while i <= 100:
        for subelement in root:
            row = {'time': subelement.get('time')}
            for child in subelement:
                for column in columns[1:]:
                    row[column] = child.get(column)
                data.append(row)
        i += 1
        print(i)
            
    # Create DataFrame from the list of dictionaries
    df = pd.DataFrame(data, columns=columns)
    
    return df
```

```python
emission_info = parse_xml_to_dataframe('../src/data/2023-04-06-17-43-39emission_info.xml', df_cols)
```
